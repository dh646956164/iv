```python
!pip install opendatasets
import opendatasets
import pandas as pd
# opendatasets.download("https://www.kaggle.com/datasets/snap/amazon-fine-food-reviews")
opendatasets.download("https://www.kaggle.com/datasets/jahaidulislam/significant-earthquake-dataset-1900-2023")

#Read the dataset 
earthquake_data = pd.read_csv('significant-earthquake-dataset-1900-2023/Significant Earthquake Dataset 1900-2023.csv')
```

```python
import altair as alt
import pandas as pd
from ipywidgets import interact
from IPython.display import display

# Disable the max row limit
alt.data_transformers.disable_max_rows()

# Preprocess the dataset
earthquake_data['Year'] = pd.to_datetime(earthquake_data['Time']).dt.year

# Function to create the chart based on the selected decade
def create_chart(decade_start, magnitude_greater_than_6=False, depth_greater_than_100=False):
    decade_end = decade_start + 9
    filtered_data = earthquake_data[(earthquake_data['Year'] >= decade_start) & (earthquake_data['Year'] <= decade_end)]

    if magnitude_greater_than_6:
        filtered_data = filtered_data[filtered_data['Mag'] > 6]

    if depth_greater_than_100:
        filtered_data = filtered_data[filtered_data['Depth'] > 100]

    # Bar chart of the number of earthquakes per year
    bar_chart = alt.Chart(filtered_data).mark_bar().encode(
        alt.X("Year:O", title="Year"),
        alt.Y("count(ID):Q", title="Number of Earthquakes"),
        tooltip=["Year:O", "count(ID):Q"]
    ).properties(
        title=f"Number of Significant Earthquakes per Year ({decade_start}-{decade_end})",
        width=600,
        height=300
    )

    # Scatter plot of earthquake depth vs. magnitude
    scatter_plot = alt.Chart(filtered_data).mark_circle().encode(
        alt.X("Depth:Q", title="Depth (km)", scale=alt.Scale(zero=False)),
        alt.Y("Mag:Q", title="Magnitude", scale=alt.Scale(zero=False)),
        color="Year:O",
        tooltip=["Depth:Q", "Mag:Q", "Year:O"]
    ).properties(
        title=f"Earthquake Depth vs. Magnitude ({decade_start}-{decade_end})",
        width=600,
        height=300
    ).interactive()  # Add zooming and panning interactivity

    display(bar_chart & scatter_plot)

# Button-based interactivity
min_year = earthquake_data['Year'].min()
max_year = earthquake_data['Year'].max()
decades = [(f"{decade_start}s", decade_start) for decade_start in range(min_year - min_year % 10, max_year - max_year % 10 + 1, 10)]

interact(create_chart, decade_start=decades, magnitude_greater_than_6=False, depth_greater_than_100=False)
```
